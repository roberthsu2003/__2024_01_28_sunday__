```python
import pandas as pd

df1 = pd.read_csv('各鄉鎮市區人口密度.csv')
df2 = df1.rename(columns={"statistic_yyy":"統計年",'site_id':'區域別','people_total':'年底人口數','area':'土地面積','population_density':'人口密度'})
df3 = df2.drop(0)
df4 = df3.reindex(index=range(1,371))
df4
```

## replace(),astype()

```python
serial1 = df4['年底人口數'].replace(to_replace='… ',value='0')
df4['年底人口數'] = serial1.astype(int)
df4.info()
```

```python
serial2 = df4['人口密度'].replace(to_replace='… ',value='0')
df4['人口密度'] = serial2.astype(int)
df4.info()
```

```python
df4
```

## sort_values()

```python
populations_df = df4.sort_values(by='年底人口數',ascending=False)
populations_df
```
