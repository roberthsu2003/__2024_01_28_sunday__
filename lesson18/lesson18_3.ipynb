```python
import pandas as pd

import pandas as pd

df1 = pd.read_csv('各鄉鎮市區人口密度.csv')
df2 = df1.rename(columns={"statistic_yyy":"統計年",'site_id':'區域別','people_total':'年底人口數','area':'土地面積','population_density':'人口密度'})
df3 = df2.drop(0)
df4 = df3.reindex(index=range(1,371))
df4
```

```python
df4.info()
```

```python
#使用for_in,只可以觀察,不可以修改
for num in df4['年底人口數']:
    print(num)
```

```python
def ch(value:str) -> int:
    try:
        return int(value)
    except:
        return 0

df4['年底人口數'] = df4['年底人口數'].map(ch)
df4.info()
```

```python
def ch1(value:int) -> str:
    if value >= 413590:
        return '人口數多'
    else:
        return "人口一般"

df4['人口數狀態'] = df4['年底人口數'].map(ch1)
df4
```
