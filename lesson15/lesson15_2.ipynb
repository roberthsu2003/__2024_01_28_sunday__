```python
import matplotlib.pyplot as plt
# 1.製作figure->畫布
fig = plt.figure()
print(type(fig))

# 2.在1.的figure上製作一個axes
ax_1 = fig.add_subplot(1, 1, 1)
print(type(ax_1))

# 3.將圖的數據儲存在axes中
X = [0, 1, 2]
Y = [3, 4, 5]
ax_1.plot(X, Y)

# 4.顯示圖
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax_1 = fig.add_subplot(1, 3, 1)
ax_1.set_title('Axes1')
ax_2 = fig.add_subplot(1, 3, 2)
ax_2.set_title('Axes2')
ax_3 = fig.add_subplot(1, 3, 3)
ax_3.set_title('Axes3')

plt.show()
```

```python
import random
import matplotlib.pyplot as plt
days = [num for num in range(1, 32)]
celsius = [round(random.uniform(28.5,35.2),ndigits=1) for _ in range(1, 32)]
figure = plt.figure(figsize=(10,5))
ax = figure.add_subplot(1,1,1)
ax.grid()
ax.plot(days,celsius,'g.-.')
ax.set_title("july weather")
ax.set_xlabel('Days')
ax.set_ylabel('Celsius')
#print(ax.get_xticks())
ax.set_xticks(days)
plt.show()
```
